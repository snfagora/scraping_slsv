## Setup

```python
# check the current working directory 

# cp *NSLVE* ./nslve

os.chdir("/home/jae/scraping_slsv/outputs/nslve")

cwd = os.getcwd()

# print the cwd 
print("Current working directory: {0}". format(cwd))
```

```python
# list files 

file_names = os.listdir(cwd)

# placeholder 

nslve_files = []

# for loop

for file in file_names:
    if fnmatch.fnmatch(file.lower(), "*nslve*"):
        nslve_files.append(file)
```

```python
# check 

nslve_files[0:4]
```

```python
# load template 

temp = []

for file in file_names:
    if fnmatch.fnmatch(file, "*json"):
        temp.append(file)
```

```python
# check
os.getcwd() + "/" + ''.join(temp)
```

## Extract Table obejcts from PDF files

```python
df = camelot.read_pdf(filepath = os.getcwd() + "/" + nslve_files[0],
                      pages = 'all')
```

```python
df = tabula.read_pdf_with_template(input_path = os.getcwd() + "/" + nslve_files[0],
                              #batch = os.getcwd(),
                              template_path = os.getcwd() + "/" + ''.join(temp),
                              stream = True,
                              guess = False,
                              pages = "all")
```
